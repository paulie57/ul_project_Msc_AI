```python
import pandas as pd
import numpy as np

# PyCaret
from pycaret.regression import RegressionExperiment

# Scikit-learn Poisson
from sklearn.linear_model import PoissonRegressor

# Statsmodels for Negative Binomial
import statsmodels.api as sm
from sklearn.base import BaseEstimator, RegressorMixin
```

```python
class NegativeBinomialRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, alpha=1.0):
        """
        alpha is the dispersion parameter in statsmodels.
        A smaller alpha ~ Poisson, a larger alpha implies more overdispersion.
        """
        self.alpha = alpha
        
    def fit(self, X, y):
        # Convert to numpy arrays
        X = np.array(X)
        y = np.array(y)
        
        # Add intercept column
        X = sm.add_constant(X, prepend=True, has_constant='add')

        # Fit Negative Binomial model
        self.model_ = sm.GLM(
            y,
            X,
            family=sm.families.NegativeBinomial(alpha=self.alpha)
        ).fit()
        
        return self
    
    def predict(self, X):
        # Predict using the fitted model
        X = np.array(X)
        X = sm.add_constant(X, prepend=True, has_constant='add')
        return self.model_.predict(X)
```

```python
exp = RegressionExperiment() 
exp.setup(merged_reg, target='fracture_count', session_id=555, train_size=0.7, categorical_features= categorical_cols, numeric_features=numeric_cols, numeric_imputation='mean', categorical_imputation='mode' ,
           preprocess=True, normalize=True , remove_outliers=True, ordinal_features=None)
```

```python
poisson_estimator = PoissonRegressor(alpha=1.0, max_iter=1000)
poisson_model = exp.create_model(estimator=poisson_estimator)
```

```python
neg_bin_estimator = NegativeBinomialRegressor(alpha=1.0)
neg_bin_model = exp.create_model(estimator=neg_bin_estimator)
```

```python
best_model = exp.compare_models(n_select=5,
    include=[poisson_model, neg_bin_model, "et", "gbr", "rf", "knn", "ada", "dt", "lasso", "ridge", "br", "lr", "lar", "huber", "llar", "en", "omp", "par", "dummy"]
)
```

```python
Huber = exp.create_model('huber')
Bayesian = exp.create_model('br')
Poisson = exp.create_model(poisson_model)
```

```python
tune_Huber = exp.tune_model(Huber)
tune_Bayesian = exp.tune_model(Bayesian)
```

```python
custom_grid = {
    'alpha': [0.5, 1.0, 2.0, 5.0]
}
tune_Poisson = exp.tune_model(estimator=poisson_model, custom_grid=custom_grid, optimize='RMSE')
```
