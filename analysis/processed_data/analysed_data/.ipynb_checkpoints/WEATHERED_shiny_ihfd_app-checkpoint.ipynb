def month_to_season(month):
    # Convert month to an integer if it isn't already.
    month = int(month)
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'unknown'

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

def run_quasipoisson_with_irr(df):
    """
    Perform a quasi-Poisson regression analysis and calculate incidence rate ratios.
    
    Parameters:
    -----------
    df : pandas DataFrame
        DataFrame containing the hip fracture data
        
    Returns:
    --------
    dict
        Dictionary containing model results and IRR table
    """
    # Specify the formula with population offset
    formula = "hip_fracture_count ~ region + weather_event:warning_severity_numeric"
    
    # Fit the regular Poisson model with offset
    poisson_model = smf.glm(
        formula=formula, 
        data=df, 
        family=sm.families.Poisson(), 
        offset=np.log(df['population'])
    )
    poisson_results = poisson_model.fit()
    
    # Calculate the dispersion parameter
    dispersion = poisson_results.deviance / poisson_results.df_resid
    print(f"Dispersion parameter: {dispersion:.4f}")
    
    # Fit the quasi-Poisson model by adjusting for overdispersion
    quasi_poisson_results = poisson_model.fit(scale=dispersion)
    
    # Print the summary of the quasi-Poisson model
    print(quasi_poisson_results.summary())
    
    # Extract parameter estimates, confidence intervals, and p-values
    params = quasi_poisson_results.params
    conf = quasi_poisson_results.conf_int()
    pvalues = quasi_poisson_results.pvalues
    
    # Calculate IRRs and confidence intervals
    irr_data = {
        'Coefficient': params,
        'IRR': np.exp(params),
        'Lower CI': np.exp(conf[0]),
        'Upper CI': np.exp(conf[1]),
        'p-value': pvalues
    }
    
    # Create a DataFrame with IRR values
    irr_df = pd.DataFrame(irr_data)
    
    # Print IRR table with nicely formatted columns
    print("\nIncidence Rate Ratios (IRR) with 95% Confidence Intervals:")
    formatted_irr = irr_df.copy()
    for col in ['IRR', 'Lower CI', 'Upper CI']:
        formatted_irr[col] = formatted_irr[col].map('{:.4f}'.format)
    formatted_irr['p-value'] = formatted_irr['p-value'].map('{:.6f}'.format)
    print(formatted_irr)
    
    # Add predicted values to the original dataframe
    df_with_pred = df.copy()
    df_with_pred['predicted'] = quasi_poisson_results.predict(offset=np.log(df['population']))
    
    # Calculate Pearson residuals
    df_with_pred['pearson_residuals'] = quasi_poisson_results.resid_pearson
    
    # Create diagnostic plots
    create_diagnostic_plots(df_with_pred, irr_df)
    
    return {
        'model': quasi_poisson_results,
        'irr_table': irr_df,
        'dispersion': dispersion,
        'df_with_predictions': df_with_pred
    }

def create_diagnostic_plots(df, irr_df):
    """
    Create diagnostic plots for the quasi-Poisson model.
    
    Parameters:
    -----------
    df : pandas DataFrame
        DataFrame with original data and model predictions
    irr_df : pandas DataFrame
        DataFrame with IRR results
    """
    # Plot 1: Actual vs Predicted
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.scatter(df['predicted'], df['hip_fracture_count'], alpha=0.5)
    max_val = max(df['predicted'].max(), df['hip_fracture_count'].max())
    plt.plot([0, max_val], [0, max_val], 'r--')
    plt.xlabel('Predicted Hip Fracture Count')
    plt.ylabel('Actual Hip Fracture Count')
    plt.title('Actual vs Predicted Hip Fracture Counts')
    
    # Plot 2: Residuals vs Predicted
    plt.subplot(1, 2, 2)
    plt.scatter(df['predicted'], df['pearson_residuals'], alpha=0.5)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel('Predicted Hip Fracture Count')
    plt.ylabel('Pearson Residuals')
    plt.title('Residuals vs Predicted Values')
    
    plt.tight_layout()
    plt.show()
    
    # Extract region effects from IRR table
    region_effects = irr_df[irr_df.index.str.contains('region')]
    
    if not region_effects.empty:
        # Sort by IRR value
        region_effects = region_effects.sort_values('IRR')
        
        # Create a horizontal bar chart for region effects
        plt.figure(figsize=(10, 6))
        
        # Create error bars for confidence intervals
        error_low = region_effects['IRR'] - region_effects['Lower CI']
        error_high = region_effects['Upper CI'] - region_effects['IRR']
        
        plt.barh(region_effects.index, region_effects['IRR'], xerr=[error_low, error_high], alpha=0.7)
        plt.axvline(x=1, color='r', linestyle='--')
        plt.xlabel('Incidence Rate Ratio (IRR)')
        plt.title('Regional Effects on Hip Fracture Risk')
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
    
    # Extract weather effect from IRR table
    weather_effect = irr_df[irr_df.index.str.contains('warning_severity')]
    
    if not weather_effect.empty:
        # Create a bar chart for weather effect
        plt.figure(figsize=(8, 4))
        
        # Create error bars for confidence intervals
        error_low = weather_effect['IRR'] - weather_effect['Lower CI']
        error_high = weather_effect['Upper CI'] - weather_effect['IRR']
        
        plt.bar(0, weather_effect['IRR'].values[0], 
                yerr=[[error_low.values[0]], [error_high.values[0]]],
                width=0.4, alpha=0.7)
        
        plt.axhline(y=1, color='r', linestyle='--')
        plt.ylabel('Incidence Rate Ratio (IRR)')
        plt.title('Effect of Weather Warning Severity on Hip Fracture Risk')
        plt.xticks([0], ['Weather Event × Warning Severity'])
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

# Example usage
results = run_quasipoisson_with_irr(df_final)
irr_table = results['irr_table']
```
