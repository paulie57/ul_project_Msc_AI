```python
def prepare_data_2022(ihfd_df, weather_df):
    """
    Prepare and merge hip fracture and weather data for 2022, focusing on the HSE Mid West region.
    This function matches each fracture incident with any weather warnings that were active at the time.
    """
    # First convert our datetime columns, removing timezone info to ensure consistent comparison
    ihfd_df['datetime'] = pd.to_datetime(ihfd_df['Adm_First_Pres_Hosp_DateTime']).dt.tz_localize(None)
    weather_df['valid_from'] = pd.to_datetime(weather_df['Valid From']).dt.tz_localize(None)
    weather_df['valid_to'] = pd.to_datetime(weather_df['Valid To']).dt.tz_localize(None)
    
    # Print diagnostic information about our initial data
    print("\nData Processing Steps:")
    print("1. Initial Data Counts:")
    print(f"Total hip fractures: {len(ihfd_df)}")
    print(f"Total weather events: {len(weather_df)}")
    
    # Filter for 2022 data
    ihfd_2022 = ihfd_df[ihfd_df['datetime'].dt.year == 2022].copy()
    weather_2022 = weather_df[weather_df['valid_from'].dt.year == 2022].copy()
    
    print("\n2. 2022 Data Counts:")
    print(f"2022 hip fractures: {len(ihfd_2022)}")
    print(f"2022 weather events: {len(weather_2022)}")
    
    # Filter for Mid West region
    midwest_weather = weather_2022[weather_2022['HSE Mid West'] == 1].copy()
    midwest_incidents = ihfd_2022[ihfd_2022['New Health Regions'] == 'HSE Mid West'].copy()
    
    print("\n3. Mid West Region Counts for 2022:")
    print(f"Mid West incidents: {len(midwest_incidents)}")
    print(f"Mid West weather events: {len(midwest_weather)}")
    
    # Add temporal features to help with analysis
    midwest_incidents['date'] = midwest_incidents['datetime'].dt.date
    midwest_incidents['month'] = midwest_incidents['datetime'].dt.month
    midwest_incidents['day_of_week'] = midwest_incidents['datetime'].dt.dayofweek
    midwest_incidents['is_weekend'] = midwest_incidents['day_of_week'].isin([5, 6]).astype(int)
    
    def get_weather_warnings(row):
        """
        For each incident, find any overlapping weather warnings.
        Returns details about the most severe warning if multiple exist.
        """
        incident_time = row['datetime']
        
        # Find all warnings that overlap with this incident time
        overlapping = midwest_weather[
            (midwest_weather['valid_from'] <= incident_time) &
            (midwest_weather['valid_to'] >= incident_time)
        ]
        
        # Print diagnostic info for every 50th incident to track progress
        if row.name % 50 == 0:
            print(f"\nProcessing incident {row.name} at {incident_time}")
            print(f"Found {len(overlapping)} overlapping warnings")
            
            if len(overlapping) > 0:
                print("Sample warning:")
                print(overlapping[['Warning Colour', 'Warning Element']].iloc[0])
        
        if len(overlapping) == 0:
            return pd.Series({
                'warning_color': 'No Active Weather Warning on Date of Hip-Fracture Incident',
                'warning_element': 'No Warning',
                'num_warnings': 0,
                'warning_duration': 0
            })
        
        # Determine the most severe warning color (Red > Orange > Yellow)
        warning_color = 'Yellow'  # default if any warning exists
        if any(overlapping['Warning Colour'].str.lower() == 'red'):
            warning_color = 'Red'
        elif any(overlapping['Warning Colour'].str.lower() == 'orange'):
            warning_color = 'Orange'
        
        # Calculate how long the warning had been active when the incident occurred
        warning_duration = min(
            (overlapping['valid_to'] - incident_time).min().total_seconds() / 3600,
            24  # Cap at 24 hours for analysis purposes
        )
        
        return pd.Series({
            'warning_color': warning_color,
            'warning_element': ', '.join(overlapping['Warning Element'].unique()),
            'num_warnings': len(overlapping),
            'warning_duration': warning_duration
        })
    
    # Apply the weather matching function and merge results
    print("\n4. Matching weather warnings to incidents...")
    weather_info = midwest_incidents.apply(get_weather_warnings, axis=1)
    midwest_incidents = pd.concat([midwest_incidents, weather_info], axis=1)
    
    # Add seasonal categorization
    midwest_incidents['season'] = pd.cut(
        midwest_incidents['month'],
        bins=[0, 3, 6, 9, 12],
        labels=['Winter', 'Spring', 'Summer', 'Autumn']
    )
    
    # Select and organize our final columns
    final_columns = [
        'datetime', 'date', 'month', 'day_of_week', 'is_weekend', 'season',
        'NOCA_FirstPresPeriodDay', 'New Health Regions',
        'warning_color', 'warning_element', 'num_warnings', 'warning_duration'
    ]
    
    final_df = midwest_incidents[final_columns].copy()
    final_df['daily_total'] = final_df.groupby('date')['datetime'].transform('count')
    
    # Print final summary
    print("\n5. Final Dataset Summary:")
    print(f"Total records: {len(final_df)}")
    print("\nWarning distribution:")
    print(final_df['warning_color'].value_counts())
    
    return final_df
```

```python
def analyze_incident_distributions(df):
    """
    Analyze how hip fractures are distributed across days with and without weather warnings.
    
    Parameters:
    df: DataFrame with hip fracture and weather warning data
    """
    # First, let's group by date and get daily statistics
    daily_summary = df.groupby('date').agg({
        'warning_color': lambda x: 'Day with Weather Warning' 
                                 if any(x != 'No Active Weather Warning on Date of Hip-Fracture Incident') 
                                 else 'Day without Weather Warning',
        'datetime': 'count'  # This counts incidents per day
    }).rename(columns={'datetime': 'daily_incidents'})
    
    # Now analyze the distributions
    warning_stats = daily_summary.groupby('warning_color').agg({
        'daily_incidents': ['count', 'sum', 'mean', 'min', 'max']
    })
    
    # Calculate the total number of incidents in each category
    total_incidents = df['warning_color'].map(
        lambda x: 'Incidents on Days with Weather Warnings' 
        if x != 'No Active Weather Warning on Date of Hip-Fracture Incident'
        else 'Incidents on Days without Weather Warnings'
    ).value_counts()
    
    print("\nDetailed Incident Analysis:")
    print("\n1. Daily Statistics:")
    print(warning_stats)
    
    print("\n2. Total Incidents by Category:")
    print(total_incidents)
    
    print("\n3. Distribution of Daily Incident Counts:")
    print("\nDays without Weather Warnings:")
    print(daily_summary[daily_summary['warning_color'] == 'Day without Weather Warning']['daily_incidents'].value_counts().sort_index())
    
    print("\nDays with Weather Warnings:")
    print(daily_summary[daily_summary['warning_color'] == 'Day with Weather Warning']['daily_incidents'].value_counts().sort_index())
    
    return daily_summary

# To use this analysis, add this after creating your final_df:
daily_analysis = analyze_incident_distributions(final_data)
```

```python
def analyze_seasonal_incident_patterns(df):
    """
    Analyze daily incident patterns by season, incorporating weather warnings.
    Uses Met Ã‰ireann's meteorological seasons with separate averages for warning and non-warning days.
    """
    def get_met_eireann_season(month):
        # Met Ã‰ireann seasons: Winter (Dec-Feb), Spring (Mar-May), Summer (Jun-Aug), Autumn (Sep-Nov)
        if month == 12 or month <= 2:
            return 'Winter'
        elif 3 <= month <= 5:
            return 'Spring'
        elif 6 <= month <= 8:
            return 'Summer'
        else:
            return 'Autumn'
    
    # Add meteorological season to the dataframe
    df['met_season'] = df['datetime'].dt.month.map(get_met_eireann_season)
    
    # Create daily summary with season information
    daily_summary = df.groupby(['date', 'met_season', 'warning_color'])['datetime'].count().reset_index(name='daily_incidents')
    
    print("Seasonal Analysis of Daily Hip Fracture Incidents:")
    
    for season in ['Winter', 'Spring', 'Summer', 'Autumn']:
        season_data = daily_summary[daily_summary['met_season'] == season]
        
        print(f"\n{'-'*20}")
        print(f"\n{season} Analysis:")
        print(f"{'-'*20}")
        
        # Analyze days without warnings
        no_warning = season_data[season_data['warning_color'] == 'No Active Weather Warning on Date of Hip-Fracture Incident']
        print("\nDays without Weather Warnings:")
        if not no_warning.empty:
            incident_dist = no_warning['daily_incidents'].value_counts().sort_index()
            total_no_warning_incidents = (incident_dist * incident_dist.index).sum()
            total_no_warning_days = incident_dist.sum()
            avg_no_warning = total_no_warning_incidents / total_no_warning_days if total_no_warning_days > 0 else 0
            
            for incidents, count in incident_dist.items():
                total_fractures = incidents * count
                print(f"{count} days had {incidents} incident(s) - Total fractures: {total_fractures}")
            print(f"Average incidents per day without warnings: {avg_no_warning:.2f}")
        
        # Initialize counters for warning days
        warning_incidents = 0
        warning_days = 0
        
        # Analyze days with different warning levels
        for warning in ['Yellow', 'Orange', 'Red']:
            warning_data = season_data[season_data['warning_color'] == warning]
            if not warning_data.empty:
                print(f"\nDays with {warning} Warnings:")
                incident_dist = warning_data['daily_incidents'].value_counts().sort_index()
                total_warning_type = (incident_dist * incident_dist.index).sum()
                total_days_type = incident_dist.sum()
                warning_incidents += total_warning_type
                warning_days += total_days_type
                
                for incidents, count in incident_dist.items():
                    total_fractures = incidents * count
                    print(f"{count} days had {incidents} incident(s) - Total fractures: {total_fractures}")
        
        # Calculate average for warning days
        if warning_days > 0:
            avg_warning = warning_incidents / warning_days
            print(f"\nAverage incidents per day with any weather warning: {avg_warning:.2f}")
        
        # Season overall summary
        total_days = len(season_data)
        total_incidents = season_data['daily_incidents'].sum()
        print(f"\nSeason Overall Summary:")
        print(f"Total days with incidents: {total_days}")
        print(f"Total incidents: {total_incidents}")
        print(f"Overall average incidents per day: {total_incidents/total_days:.2f}")
        
        # Percentage of days with warnings
        warning_percentage = (warning_days / total_days) * 100 if total_days > 0 else 0
        print(f"Percentage of days with weather warnings: {warning_percentage:.1f}%")

# To use:
analyze_seasonal_incident_patterns(final_data)
```
