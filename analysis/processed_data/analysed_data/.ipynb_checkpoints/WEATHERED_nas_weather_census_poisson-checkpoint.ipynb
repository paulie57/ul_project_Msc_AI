```python
def aggregate_weather_data(df_weather, keep_columns=None):
    """
    Aggregate weather data by region, date, weather_type, and warning_severity while
    preserving specified additional columns.
    
    Args:
        df_weather: DataFrame containing weather data
        keep_columns: List of additional columns to preserve (beyond the groupby columns)
        
    Returns:
        Aggregated DataFrame with one row per unique region-date-weather_type-warning_severity combination
    """
    # Define default columns to keep if none provided
    if keep_columns is None:
        keep_columns = [
            'counties_in_region', 	
            'county_weight'	
         ]
    
    # Define all columns to group by
    group_cols = ['region', 'date', 'warning_phenomenon', 'warning_severity']
    
    # Define aggregation functions for numeric and other columns
    agg_dict = {
        'warning_severity_numeric': 'max'
    }
    
    # Add aggregation functions for additional columns to preserve
    for col in keep_columns:
        if col in df_weather.columns:
            # For string columns, take the first value (assuming they're the same within groups)
            if df_weather[col].dtype == 'object':
                agg_dict[col] = 'first'
            # For boolean or integer flags (like has_multiple_events, warning_upgraded)
            elif df_weather[col].dtype in ['bool', 'int64', 'int32']:
                agg_dict[col] = 'max'
            # For numeric columns, take the max
            elif pd.api.types.is_numeric_dtype(df_weather[col]):
                agg_dict[col] = 'max'
            # For datetime columns
            elif pd.api.types.is_datetime64_dtype(df_weather[col]):
                agg_dict[col] = 'max'
            # Default to first value for any other types
            else:
                agg_dict[col] = 'first'
    
    # Group by the specified columns and aggregate
    weather_agg = df_weather.groupby(group_cols, as_index=False).agg(agg_dict)
    
    # Add the weather flag column (set to 1 if there is any event)
    weather_agg['weather_event'] = 1
    
    return weather_agg
```

```python
# Add month number
df_with_pop['month'] = pd.DatetimeIndex(df_with_pop['date']).month

# Create seasons (meteorological seasons)
def assign_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:  # months 9, 10, 11
        return 'Autumn'

df_with_pop['season'] = df_with_pop['month'].apply(assign_season)

# Create dummy variables for seasons
df_with_pop['is_winter'] = (df_with_pop['season'] == 'Winter').astype(int)
df_with_pop['is_spring'] = (df_with_pop['season'] == 'Spring').astype(int)
df_with_pop['is_summer'] = (df_with_pop['season'] == 'Summer').astype(int)
df_with_pop['is_autumn'] = (df_with_pop['season'] == 'Autumn').astype(int)
```

```python
def extract_coefficients(model_summary_text):
    # Create a dictionary to store the extracted data
    coef_data = {}
    
    # Split the text by lines
    lines = model_summary_text.split('\n')
    
    # Find the line that starts the coefficient section
    start_idx = 0
    for i, line in enumerate(lines):
        if '-' * 20 in line and 'Coef.' in line:
            start_idx = i + 1
            break
    
    # Extract coefficients, standard errors, z-values, and p-values
    for line in lines[start_idx:]:
        # Skip separator lines or empty lines
        if '=' * 20 in line or line.strip() == '':
            continue
            
        # Split by whitespace, but join the variable name parts
        parts = line.strip().split()
        if len(parts) < 6:  # Need at least variable name and numeric values
            continue
            
        # Find where numeric values start
        num_start_idx = 0
        for i, part in enumerate(parts):
            # Try to convert to float to find the first numeric value
            try:
                float(part)
                num_start_idx = i
                break
            except ValueError:
                continue
        
        if num_start_idx == 0:
            continue  # No numeric values found
            
        # Join the variable name parts
        var_name = ' '.join(parts[:num_start_idx]).strip()
        
        # Extract the numeric values
        try:
            coef = float(parts[num_start_idx])
            std_err = float(parts[num_start_idx + 1])
            z = float(parts[num_start_idx + 2])
            p_value = float(parts[num_start_idx + 3])
            ci_lower = float(parts[num_start_idx + 4])
            ci_upper = float(parts[num_start_idx + 5])
            
            coef_data[var_name] = {
                'coef': coef,
                'std_err': std_err,
                'z': z,
                'p_value': p_value,
                'ci_lower': ci_lower,
                'ci_upper': ci_upper
            }
        except (ValueError, IndexError):
            # Skip if conversion fails or not enough values
            continue
    
    # Check if the coefficients were extracted correctly
    print(f"Extracted {len(coef_data)} coefficients:")
    for i, (key, value) in enumerate(list(coef_data.items())[:5]):
        print(f"  {key}: {value['coef']}")
    if len(coef_data) > 5:
        print(f"  ... and {len(coef_data) - 5} more")
    
    return coef_data

# Model summary text (paste the complete model summary here as a string)
model_summary = """
Mixed Linear Model Regression Results
============================================================================================================================
Model:                                   MixedLM                      Dependent Variable:                      callout_count
No. Observations:                        68989                        Method:                                  REML         
No. Groups:                              6                            Scale:                                   1.2628       
Min. group size:                         7525                         Log-Likelihood:                          -106054.6553 
Max. group size:                         13607                        Converged:                               Yes          
Mean group size:                         11498.2                                                                            
----------------------------------------------------------------------------------------------------------------------------
                                                                                 Coef.  Std.Err.    z    P>|z| [0.025 0.975]
----------------------------------------------------------------------------------------------------------------------------
Intercept                                                                        11.757    7.486   1.570 0.116 -2.916 26.431
C(year)[T.2017]                                                                   0.110    0.021   5.330 0.000  0.070  0.151
C(year)[T.2018]                                                                   0.251    0.025   9.894 0.000  0.201  0.301
C(year)[T.2019]                                                                   0.360    0.032  11.333 0.000  0.298  0.423
C(year)[T.2020]                                                                   0.355    0.039   9.095 0.000  0.278  0.431
C(year)[T.2021]                                                                   0.424    0.046   9.113 0.000  0.333  0.515
C(year)[T.2022]                                                                   0.617    0.054  11.402 0.000  0.511  0.723
C(year)[T.2023]                                                                   0.713    0.062  11.487 0.000  0.591  0.834
C(month)[T.2]                                                                     0.005    0.021   0.221 0.825 -0.037  0.047
C(month)[T.3]                                                                    -0.010    0.023  -0.422 0.673 -0.054  0.035
C(month)[T.4]                                                                    -0.009    0.022  -0.426 0.670 -0.053  0.034
C(month)[T.5]                                                                     0.038    0.022   1.749 0.080 -0.005  0.081
C(month)[T.6]                                                                     0.092    0.022   4.146 0.000  0.049  0.135
C(month)[T.7]                                                                     0.104    0.022   4.744 0.000  0.061  0.148
C(month)[T.8]                                                                     0.135    0.022   6.210 0.000  0.093  0.178
C(month)[T.9]                                                                     0.115    0.022   5.263 0.000  0.072  0.159
C(month)[T.10]                                                                    0.081    0.022   3.660 0.000  0.037  0.124
C(month)[T.11]                                                                    0.087    0.022   3.947 0.000  0.044  0.130
C(month)[T.12]                                                                    0.239    0.020  11.851 0.000  0.200  0.279
C(fall_category_group)[T.17A03_injuries_excluding_distal]                        -0.695    0.019 -36.222 0.000 -0.733 -0.658
C(fall_category_group)[T.17A04_public_assistance]                                -0.691    0.020 -34.659 0.000 -0.730 -0.651
C(fall_category_group)[T.17B00_bravo_override]                                   -0.185    0.016 -11.514 0.000 -0.216 -0.153
C(fall_category_group)[T.17B01_possibly_dangerous]                                1.017    0.015  68.832 0.000  0.988  1.046
C(fall_category_group)[T.17D04_not_alert]                                        -0.202    0.016 -12.560 0.000 -0.234 -0.171
C(fall_category_group)[T.30A03_traumatic_non_recent]                             -0.720    0.020 -35.687 0.000 -0.760 -0.681
weather_event:C(season)[Autumn]                                                   0.004    0.035   0.117 0.907 -0.065  0.073
weather_event:C(season)[Spring]                                                   0.046    0.037   1.259 0.208 -0.026  0.118
weather_event:C(season)[Summer]                                                   0.031    0.035   0.888 0.375 -0.038  0.101
weather_event:C(season)[Winter]                                                   0.057    0.032   1.745 0.081 -0.007  0.120
C(fall_category_group)[17A02_not_dangerous_proximal]:warning_severity_numeric    -0.062    0.030  -2.084 0.037 -0.119 -0.004
C(fall_category_group)[17A03_injuries_excluding_distal]:warning_severity_numeric -0.064    0.035  -1.845 0.065 -0.133  0.004
C(fall_category_group)[17A04_public_assistance]:warning_severity_numeric         -0.080    0.036  -2.206 0.027 -0.151 -0.009
C(fall_category_group)[17B00_bravo_override]:warning_severity_numeric             0.056    0.029   1.918 0.055 -0.001  0.113
C(fall_category_group)[17B01_possibly_dangerous]:warning_severity_numeric         0.019    0.028   0.672 0.502 -0.036  0.073
C(fall_category_group)[17D04_not_alert]:warning_severity_numeric                 -0.006    0.030  -0.201 0.841 -0.065  0.053
C(fall_category_group)[30A03_traumatic_non_recent]:warning_severity_numeric      -0.105    0.037  -2.872 0.004 -0.177 -0.033
log_population                                                                   -0.860    0.626  -1.373 0.170 -2.087  0.367
Group Var                                                                         0.330    0.256                            
============================================================================================================================
"""

# Extract coefficients from the model summary
coefficients = extract_coefficients(model_summary)

# Set the style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("talk")


# 1. Year effects visualization
years = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
year_coefs = [coefficients[f'C(year)[T.{year}]']['coef'] for year in years]
year_ci_lower = [coefficients[f'C(year)[T.{year}]']['ci_lower'] for year in years]
year_ci_upper = [coefficients[f'C(year)[T.{year}]']['ci_upper'] for year in years]

years_df = pd.DataFrame({
    'Year': years,
    'Coefficient': year_coefs,
    'Lower_CI': year_ci_lower,
    'Upper_CI': year_ci_upper
})

plt.figure(figsize=(12, 6))
plt.plot(years_df['Year'], years_df['Coefficient'], marker='o', linewidth=2, color='#1f77b4')
plt.fill_between(years_df['Year'], years_df['Lower_CI'], years_df['Upper_CI'], alpha=0.2, color='#1f77b4')
plt.title('Yearly Trend in Hip Fracture Ambulance Callouts (2017-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Coefficient (Effect Size)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(years)
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.7)
for i, year in enumerate(years):
    plt.annotate(f"{year_coefs[i]:.3f}", (year, year_coefs[i]), 
                 textcoords="offset points", xytext=(0,10), ha='center')
plt.tight_layout()
plt.savefig('/home/paulharford/college/project/ul_project_Msc_AI/analysis/processed_data/visual_outputs/nas_weather/yearly_trend.png', dpi=300, bbox_inches='tight')
plt.show()

# 2. Month effects visualization
months = list(range(1, 13))
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# January is the reference category (coef = 0)
month_coefs = [0.0]  
month_p = [1.0]  # Reference category p-value set to 1

# Add coefficients for months 2-12
for i in range(2, 13):
    key = f'C(month)[T.{i}]'
    if key in coefficients:
        month_coefs.append(coefficients[key]['coef'])
        month_p.append(coefficients[key]['p_value'])

months_df = pd.DataFrame({
    'Month': month_names,
    'Coefficient': month_coefs,
    'Significant': [p < 0.05 for p in month_p]
})

plt.figure(figsize=(14, 6))
bars = plt.bar(months_df['Month'], months_df['Coefficient'], 
       color=[sns.color_palette("rocket_r")[0] if sig else 'lightgray' for sig in months_df['Significant']])
plt.title('Monthly Patterns in Hip Fracture Ambulance Callouts', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Coefficient (Effect Size)', fontsize=14)
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.7)
plt.grid(True, axis='y', alpha=0.3)

# Add coefficient values above each bar
for bar in bars:
    height = bar.get_height()
    if height > 0:
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'{height:.3f}', ha='center', va='bottom')
    else:
        plt.text(bar.get_x() + bar.get_width()/2., height - 0.02,
                f'{height:.3f}', ha='center', va='top')

plt.tight_layout()
plt.savefig('monthly_pattern.png', dpi=300, bbox_inches='tight')
plt.show()

# 3. Extract fall categories base effects
fall_categories = ['17A02_not_dangerous_proximal', '17A03_injuries_excluding_distal', 
                   '17A04_public_assistance', '17B00_bravo_override',
                   '17B01_possibly_dangerous', '17D04_not_alert',
                   '30A03_traumatic_non_recent']
fall_cat_short = ['17A02', '17A03', '17A04', '17B00', '17B01', '17D04', '30A03']

# Base effects (17A02 is reference category with coef = 0)
fall_cat_base = [0.0]  # First category is reference
fall_cat_keys = [
    'C(fall_category_group)[T.17A03_injuries_excluding_distal]',
    'C(fall_category_group)[T.17A04_public_assistance]',
    'C(fall_category_group)[T.17B00_bravo_override]',
    'C(fall_category_group)[T.17B01_possibly_dangerous]',
    'C(fall_category_group)[T.17D04_not_alert]',
    'C(fall_category_group)[T.30A03_traumatic_non_recent]'
]

for key in fall_cat_keys:
    if key in coefficients:
        fall_cat_base.append(coefficients[key]['coef'])

# Extract weather severity interaction effects
weather_keys = [
    'C(fall_category_group)[17A02_not_dangerous_proximal]:warning_severity_numeric',
    'C(fall_category_group)[17A03_injuries_excluding_distal]:warning_severity_numeric',
    'C(fall_category_group)[17A04_public_assistance]:warning_severity_numeric',
    'C(fall_category_group)[17B00_bravo_override]:warning_severity_numeric',
    'C(fall_category_group)[17B01_possibly_dangerous]:warning_severity_numeric',
    'C(fall_category_group)[17D04_not_alert]:warning_severity_numeric',
    'C(fall_category_group)[30A03_traumatic_non_recent]:warning_severity_numeric'
]

weather_coefs = []
weather_p = []

for key in weather_keys:
    if key in coefficients:
        weather_coefs.append(coefficients[key]['coef'])
        weather_p.append(coefficients[key]['p_value'])

weather_df = pd.DataFrame({
    'Category': fall_categories,
    'Short_Name': fall_cat_short,
    'Weather_Effect': weather_coefs,
    'Significant': [p < 0.05 for p in weather_p]
})

# Fall categories visualization
plt.figure(figsize=(14, 7))
bars = plt.bar(weather_df['Short_Name'], weather_df['Weather_Effect'], 
       color=[sns.color_palette()[0] if sig else 'lightgray' for sig in weather_df['Significant']])
plt.title('Impact of Weather Warning Severity on Different Fall Categories', fontsize=16)
plt.xlabel('Fall Category', fontsize=14)
plt.ylabel('Effect of Increased Weather Warning Severity', fontsize=14)
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.7)
plt.grid(True, axis='y', alpha=0.3)

# Add coefficient values above/below each bar
for bar in bars:
    height = bar.get_height()
    if height > 0:
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.005,
                f'{height:.3f}', ha='center', va='bottom')
    else:
        plt.text(bar.get_x() + bar.get_width()/2., height - 0.01,
                f'{height:.3f}', ha='center', va='top')

plt.tight_layout()
plt.savefig('/home/paulharford/college/project/ul_project_Msc_AI/analysis/processed_data/visual_outputs/nas_weather/fall_weather_interaction.png', dpi=300, bbox_inches='tight')
plt.show()

# 4. Season weather event interaction
seasons = ['Autumn', 'Spring', 'Summer', 'Winter']
season_coefs = []
season_p = []

for season in seasons:
    key = f'weather_event:C(season)[{season}]'
    if key in coefficients:
        season_coefs.append(coefficients[key]['coef'])
        season_p.append(coefficients[key]['p_value'])

season_df = pd.DataFrame({
    'Season': seasons,
    'Weather_Effect': season_coefs,
    'Significant': [p < 0.1 for p in season_p]  # Using p<0.1 for marginal significance
})

plt.figure(figsize=(10, 6))
bars = plt.bar(season_df['Season'], season_df['Weather_Effect'], 
       color=[sns.color_palette()[0] if sig else 'lightgray' for sig in season_df['Significant']])
plt.title('Seasonal Effects of Weather Events on Callouts', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Weather Event Effect', fontsize=14)
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.7)
plt.grid(True, axis='y', alpha=0.3)

# Add coefficient values above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.002,
            f'{height:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('/home/paulharford/college/project/ul_project_Msc_AI/analysis/processed_data/visual_outputs/nas_weather/seasonal_weather_effects.png', dpi=300, bbox_inches='tight')
plt.show()

# 5. Combined visualization of base effects and weather interactions
# Create dataframes with complete data for both effects
fall_cat_df = pd.DataFrame({
    'Category': fall_categories,
    'Short_Name': fall_cat_short,
    'Base_Effect': fall_cat_base
})

# Creating a figure with two y-axes
fig, ax1 = plt.subplots(figsize=(14, 8))
ax2 = ax1.twinx()

# Bar plot for base effects
bars1 = ax1.bar(np.arange(len(fall_cat_df)) - 0.2, fall_cat_df['Base_Effect'], width=0.4, 
               color='#1f77b4', label='Base Effect')
ax1.set_ylabel('Base Effect (Primary Coefficient)', fontsize=14, color='#1f77b4')
ax1.tick_params(axis='y', labelcolor='#1f77b4')
ax1.set_ylim(-1.0, 1.5)

# Bar plot for weather interaction effects
bars2 = ax2.bar(np.arange(len(weather_df)) + 0.2, weather_df['Weather_Effect'], width=0.4, 
               color=['#ff7f0e' if sig else 'lightgray' for sig in weather_df['Significant']], 
               label='Weather Interaction')
ax2.set_ylabel('Weather Effect (Interaction Coefficient)', fontsize=14, color='#ff7f0e')
ax2.tick_params(axis='y', labelcolor='#ff7f0e')
ax2.set_ylim(-0.15, 0.15)

# X-axis settings
ax1.set_xticks(np.arange(len(fall_cat_df)))
ax1.set_xticklabels(fall_cat_df['Short_Name'], rotation=45)
ax1.set_xlabel('Fall Category', fontsize=14)

# Title and grid
plt.title('Base Effects of Fall Categories and Their Weather Warning Interactions', fontsize=16)
ax1.grid(True, axis='y', alpha=0.3)
ax1.axhline(y=0, color='gray', linestyle='--', alpha=0.7)

# Legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

# Add coefficient values
for i, (bar1, bar2) in enumerate(zip(bars1, bars2)):
    height1 = bar1.get_height()
    height2 = bar2.get_height()
    
    # Base effect values
    if height1 >= 0:
        ax1.text(bar1.get_x() + bar1.get_width()/2., height1 + 0.05,
                f'{height1:.2f}', ha='center', va='bottom', color='#1f77b4')
    else:
        ax1.text(bar1.get_x() + bar1.get_width()/2., height1 - 0.05,
                f'{height1:.2f}', ha='center', va='top', color='#1f77b4')
    
    # Weather effect values
    if height2 >= 0:
        ax2.text(bar2.get_x() + bar2.get_width()/2., height2 + 0.01,
                f'{height2:.3f}', ha='center', va='bottom', color='#ff7f0e')
    else:
        ax2.text(bar2.get_x() + bar2.get_width()/2., height2 - 0.01,
                f'{height2:.3f}', ha='center', va='top', color='#ff7f0e')

plt.tight_layout()
plt.savefig('/home/paulharford/college/project/ul_project_Msc_AI/analysis/processed_data/visual_outputs/nas_weather/combined_effects.png', dpi=300, bbox_inches='tight')
plt.show()

# 6. Heatmap visualization of predicted impacts
# Create a matrix of values for different fall categories and weather warning levels
warning_levels = [0, 1, 2, 3]  # Weather warning severity levels

# Initialize a matrix to hold the effects
effect_matrix = np.zeros((len(fall_cat_df), len(warning_levels)))

# Calculate effects (base effect + warning_level * interaction)
for i in range(len(fall_cat_df)):
    base_effect = fall_cat_df['Base_Effect'].iloc[i]
    weather_effect = weather_df['Weather_Effect'].iloc[i]
    
    for j, level in enumerate(warning_levels):
        effect_matrix[i, j] = base_effect + (level * weather_effect)

# Create a heatmap
plt.figure(figsize=(12, 8))
cmap = sns.diverging_palette(240, 10, as_cmap=True)
ax = sns.heatmap(effect_matrix, annot=True, fmt=".2f", cmap=cmap, center=0,
                xticklabels=warning_levels, yticklabels=fall_cat_df['Short_Name'])
plt.title('Predicted Effects by Fall Category and Weather Warning Level', fontsize=16)
plt.xlabel('Weather Warning Severity Level', fontsize=14)
plt.ylabel('Fall Category', fontsize=14)
plt.tight_layout()
plt.savefig('/home/paulharford/college/project/ul_project_Msc_AI/analysis/processed_data/visual_outputs/nas_weather/effect_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("All visualizations have been created successfully.")
```
